**논리회로 구현**

```python
def AND(x1, x2) :
  w1, w2, theta = 0.5, 0.5, 0.7
  tmp = x1*w1 + x2*w2
  if tmp <= theta :
    return 0
  elif tmp > theta :
    return 1  
```

```python
print(AND(0,0))
print(AND(1,0))
print(AND(0,1))
print(AND(1,1))
```

**가중치와 편향을 도입한 AND 게이트**

```python
x1 = 1
x2 = 1

def AND(x1, x2) :
  x = np.array([x1,x2])
  w = np.array([0.5, 0.5])
  b = -0.7
  tmp = np.sum(w*x) +b

  if tmp <= 0 :
    return 0
  else :
    return 1    
```

```python
x1 = 0
x2 = 1

def NAND(x1, x2) :
  x = np.array([x1,x2])
  w = np.array([-0.5, -0.5])
  b =0.7
  tmp = np.sum(w*x) +b

  if tmp <= 0 :
    return 0
  else :
    return 1    
```

```python
x1 = 0
x2 = 1

def OR(x1, x2) :
  x = np.array([x1,x2])
  w = np.array([0.5, 0.5])
  b = -0.2
  tmp = np.sum(w*x) +b

  if tmp <= 0 :
    return 0
  else :
    return 1    
```

```python
print(AND(x1,x2))
```

**XOR 게이트 구현**

```python
 def XOR(x1, x2) :
   s1 = NAND(x1, x2)
   s2 = OR(x1, x2)
   y = AND(s1,s2)
   return y

print(XOR(0,0))
```
