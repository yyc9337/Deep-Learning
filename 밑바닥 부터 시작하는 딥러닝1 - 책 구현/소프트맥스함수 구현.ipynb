```python
import numpy as np
```

```python
a = np.array([0.3, 2.9, 4.0])

exp_a = np.exp(a)
print(exp_a)
```

```python
sum_exp_a = np.sum(exp_a)
print(sum_exp_a)
```

```python
y = exp_a / sum_exp_a
print(y)
```

오버플로우 문제를 주의해야함

```python
def softmax(a) :
  c = np.max(a)
  exp_a = np.exp(a - c) # 오버플로우에 대한 대책
  sum_exp_a = np.sum(exp_a)
  y = exp_a / sum_exp_a

  return y
```

```python
softmax(a)
```
